```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sk
from plotnine import *

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import SpectralClustering

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from scipy import linalg
from sklearn.datasets import make_sparse_spd_matrix
## GraphLassCV is GraphicalLassoCV in newer versions of scikit-learn.
from sklearn.covariance import GraphLassoCV, ledoit_wolf

from mpl_toolkits import mplot3d
```

```python
genome_x_mut_df = pd.read_csv("../results/GxM_matrix.csv",index_col=0)
''' map samples to environment.'''
sample_to_env = {}
DM0_DM25_metadata = pd.read_csv('../data/rohan-formatted/populations-and-clones.csv')
LTEE_50K_metadata = pd.read_csv('../data/rohan-formatted/LTEE-50K-clones.csv')

for i,row in DM0_DM25_metadata.iterrows():
    sample_to_env[row.Name] = row.Environment

for i,row in LTEE_50K_metadata.iterrows():
    sample_to_env[row.Name] = row.Environment

''' make a list that will be a column of true environment labels'''
env_labels = []
for i,row in genome_x_mut_df.iterrows():
    env_labels.append(sample_to_env[i])
    
    
# Scaling mean = 0, var = 1
standardized_genome_x_mut_df = pd.DataFrame(genome_x_mut_df,#pd.DataFrame(StandardScaler().fit_transform(genome_x_mut_df), 
                           index = genome_x_mut_df.index,
                           columns = genome_x_mut_df.columns)
```

## Instead of PCA, use Linear Discriminant Analysis.

```python
lda = LinearDiscriminantAnalysis(n_components=2)

env_to_color = {'DM0':0,'DM25':1,'LTEE':2}

my_target = [env_to_color[x] for x in env_labels]

test = lda.fit(genome_x_mut_df, my_target).transform(genome_x_mut_df)

LDA_plot_df = pd.DataFrame({'x':test[:,0],
                            'y':test[:,1],
                            'environment':env_labels,
                           'genome':[genome.name for dummy,genome in standardized_genome_x_mut_df.iterrows()]})

(ggplot(LDA_plot_df,aes(x='x',y='y',color='environment',label='genome')) + 
 geom_text(size=8) + 
 theme_classic())
```

```python
lda2 = LinearDiscriminantAnalysis(n_components=3)

env_to_color = {'DM0':0,'DM25':1,'LTEE':2}

my_target = [env_to_color[x] for x in env_labels]

test = lda2.fit(genome_x_mut_df, my_target).transform(genome_x_mut_df)

LDA_plot_3D = pd.DataFrame({'x':test[:,0],
                            'y':test[:,1],
                            'z':test[:,2],
                            'environment':env_labels,
                           'genome':[genome.name for dummy,genome in standardized_genome_x_mut_df.iterrows()]})

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(LDA_plot3D['x'], LDA_plot3D['y'], LDA_plot3D['z'], c=[env_to_color[x] for x in LDA_plot3D['environment']],cmap='viridis')
```

### Plot spectral clustering of vectors and true labels on top of PCA.

```python
pca = PCA(n_components=3).fit(genome_x_mut_df)
pca_components = pca.transform(genome_x_mut_df)

#pca = PCA(n_components=3).fit(standardized_genome_x_mut_df)
#pca_components = pca.transform(standardized_genome_x_mut_df)

spectral_clust = SpectralClustering(n_clusters=3,affinity='nearest_neighbors').fit(standardized_genome_x_mut_df)
cluster_assignment = spectral_clust.fit_predict(standardized_genome_x_mut_df)

PCA_plot_df = pd.DataFrame({'x':pca_components[:,0],
                            'y':pca_components[:,1],
                            'environment':env_labels,
                            'spectral_clustering':cluster_assignment,
                           'genome':[genome.name for dummy,genome in standardized_genome_x_mut_df.iterrows()]})

(ggplot(PCA_plot_df,aes(x='x',y='y',color='environment',shape='spectral_clustering')) + 
 geom_point() + 
 theme_classic())
```

```python
import numpy as np
from contrastive import CPCA

N = 400; D = 30; gap=3
# In B, all the data pts are from the same distribution, which has different variances in three subspaces.
B = np.zeros((N, D))
B[:,0:10] = np.random.normal(0,10,(N,10))
B[:,10:20] = np.random.normal(0,3,(N,10))
B[:,20:30] = np.random.normal(0,1,(N,10))


# In A there are four clusters.
A = np.zeros((N, D))
A[:,0:10] = np.random.normal(0,10,(N,10))
# group 1
A[0:100, 10:20] = np.random.normal(0,1,(100,10))
A[0:100, 20:30] = np.random.normal(0,1,(100,10))
# group 2
A[100:200, 10:20] = np.random.normal(0,1,(100,10))
A[100:200, 20:30] = np.random.normal(gap,1,(100,10))
# group 3
A[200:300, 10:20] = np.random.normal(2*gap,1,(100,10))
A[200:300, 20:30] = np.random.normal(0,1,(100,10))
# group 4
A[300:400, 10:20] = np.random.normal(2*gap,1,(100,10))
A[300:400, 20:30] = np.random.normal(gap,1,(100,10))
A_labels = [0]*100+[1]*100+[2]*100+[3]*100

cpca = CPCA(standardize=False)
cpca.fit_transform(A, B, plot=True, active_labels=A_labels)
```

### Fit UGM to data by inferring a sparse inverse covariance matrix.
### https://scikit-learn.org/stable/auto_examples/covariance/plot_sparse_cov.html

```python
test = genome_x_mut_df.filter(items=['yhiO', 'gltA','ygjA','aceB','dcuS','priC','fadL',
                                     'cls','yobG','menC','mglB','nlpI','fadE','dinI','mdoB','rpoB'])
test
```

```python
model = GraphLassoCV(cv=5)
model.fit(test)
cov_ = model.covariance_
prec_ = model.precision_

lw_cov_, _ = ledoit_wolf(test)
lw_prec_ = linalg.inv(lw_cov_)
```
